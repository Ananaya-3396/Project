```python
df = pd.read_csv("Bengaluru_House_Data.csv")
df
```

```python
df = df.drop(['society','availability','size'],axis='columns')
```

```python
df.info()
```

```python
df = df.dropna()
df.head()
```

```python
df.info()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df["area_type"] = label_encoder.fit_transform(df["area_type"])
df["location"] = label_encoder.fit_transform(df["location"])
```

```python
df.info()
```

```python
df['total_sqft']
```

```python
df = df.dropna(subset=['total_sqft'])
df.head()
```

```python
# Define a function to convert range strings to numerical values
def convert_range_to_num(s):
    tokens = s.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(s)
    except:
        return None

# Apply the function to the 'total_sqft' column and create a new column 'total_sqft_numeric'
df['total_sqft_numeric'] = df['total_sqft'].apply(convert_range_to_num)
```

```python
# Drop rows with non-numerical values in the new column
df = df.dropna(subset=['total_sqft_numeric'])

# Convert the 'total_sqft_numeric' column to integer data type
df['total_sqft_numeric'] = df['total_sqft_numeric'].astype(int)
```

```python
df.head()
```

```python
msk = np.random.rand(len(df))<=0.85
train=df[msk]   
test=df[~msk]
```

```python
x = np.asanyarray(train[["total_sqft_numeric"]])
```

```python
y = np.asanyarray(train[["price"]])
```

```python
from sklearn import linear_model
regr= linear_model.LinearRegression()
regr
```

```python
train_x = np.asanyarray(train[["total_sqft_numeric"]])
train_y = np.asanyarray(train[["price"]])
```

```python
test_x = np.asanyarray(test[["total_sqft_numeric"]])  
test_y = np.asanyarray(test[["price"]])
```

```python
regr.fit(train_x,train_y)
```

```python
print("Coefficient : " , regr.coef_)
print("Intercept : " , regr.intercept_)
```

```python
predicted_y = regr.predict(test_x)
predicted_y
```

```python
from sklearn.metrics import r2_score 
print(f"R2-score:{r2_score(test_y , predicted_y)}")
```
