```python
data = pd.read_csv("F:\insideAIML\DDOS_CLEANED1.csv")
data.head()
```

### Applying Co-relation and Co-variance

```python
cor_matrix = data.corr().abs()
cor_matrix
```

```python
import numpy as np
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.85)]
print(to_drop)
```

```python
data.drop(['pkSeqID', 'flgs_number', 'proto', 'proto_number', 'bytes', 'state', 'state_number', 'ltime', 'max', 'spkts', 'dpkts', 'sbytes', 'dbytes', 'TnBPDstIP', 'TnP_PDstIP', 'TnP_Per_Dport', 'AR_P_Proto_P_DstIP', 'AR_P_Proto_P_Dport', 'attack', 'category', 'subcategory'], axis = 1,inplace = True)
data.head()
```

```python
data.info()
```

```python
data.drop(["Unnamed: 0","stime"], axis = 1, inplace = True)
data.head()
```

```python
data['flgs'] = pd.to_numeric(data['flgs'], downcast='float')
data['pkts'] = pd.to_numeric(data['pkts'], downcast='float')
data['seq'] = pd.to_numeric(data['seq'], downcast='float')
data['TnP_PSrcIP'] = pd.to_numeric(data['TnP_PSrcIP'], downcast='float')
data['Pkts_P_State_P_Protocol_P_DestIP'] = pd.to_numeric(data['Pkts_P_State_P_Protocol_P_DestIP'], downcast='float')
data.head()
```

```python
data.info()
```

```python
data.drop(["N_IN_Conn_P_DstIP","N_IN_Conn_P_SrcIP","Pkts_P_State_P_Protocol_P_SrcIP"], axis = 1, inplace = True)
data.head()
```

```python
data.drop("Pkts_P_State_P_Protocol_P_DestIP", axis = 1, inplace = True)
data.head()
```

```python
from sklearn.feature_selection import VarianceThreshold
var_thres = VarianceThreshold(threshold=0.20)
var_thres.fit(data)
```

```python
var_thres.get_support()
```

#### Now we need to remove category because it is the only one with less that 20% variance. Over here no column haeve less that 20% variance, so we will remove none.

```python
x = data.drop("TnP_PSrcIP",axis = 1)
y = data["TnP_PSrcIP"]
print(x.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.1, random_state = 10)
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

## Applying logistic regression

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(solver = "newton-cg")
LR.fit(x_train,y_train) #training 

y_pred1 = LR.predict(x_test)

from sklearn.metrics import f1_score
print("Logistic Regression Accuracy is : ", f1_score(y_test, y_pred1)*100,"%")
```

## Applying Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
LoanTree = DecisionTreeClassifier(criterion="entropy", max_depth = 5)  
LoanTree.fit(x_train,y_train)

y_pred3 = LoanTree.predict(x_test)

from sklearn.metrics import f1_score
print("Decision Trees accuracy is: ",f1_score(y_test,y_pred3)*100,"%")
```

## Applying Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000,criterion="entropy")  
rf.fit(x_train,y_train)

y_pred4 = rf.predict(x_test)

from sklearn.metrics import f1_score
print("Random Forest accuracy is: ",f1_score(y_test,y_pred4)*100,"%")
```
