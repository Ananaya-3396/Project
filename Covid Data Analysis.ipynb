```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("covid.csv")
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
india_cases = data[data["location"]=="India"]
india_cases
```

```python
pakistan_cases = data[data["location"]=="Pakistan"]
pakistan_cases
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib.pyplot as mp
```

```python
sns.set(rc={'figure.figsize':(25,20)})
sns.lineplot(x="date",y="total_cases",data=india_cases.tail(25))
plt.show()
```

```python
sns.set(rc={'figure.figsize':(25,20)})
sns.lineplot(x="date",y="total_cases",data=pakistan_cases.tail(25))
plt.show()
```

```python
india_cases.head().plot(x="extreme_poverty", y="cvd_death_rate", kind="bar", figsize=(10, 10))
mp.show()
```

```python
#Getting latest data
last_day_cases=data[data["date"]=="2020-05-24"]
last_day_cases
```

```python
#Sorting data w.r.t total_cases
max_cases_country=last_day_cases.sort_values(by="total_cases",ascending=False)
max_cases_country.head()
```

```python
# least 5 countries
max_cases_country.tail()
```

```python
#Making bar-plot for countries with top cases
sns.barplot(x="location",y="total_cases",data=max_cases_country[1:10],hue="location")
plt.show()
```
