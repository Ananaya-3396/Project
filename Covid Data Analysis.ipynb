```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("covid.csv")
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
india_cases = data[data["location"]=="India"]
india_cases
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib.pyplot as mp
```

```python
sns.set(rc={'figure.figsize':(25,20)})
sns.lineplot(x="date",y="total_cases",data=india_cases.head(25))
plt.show()
```

```python
sns.set(rc={'figure.figsize':(25,20)})
sns.lineplot(x="date",y="total_cases",data=india_cases.tail(25))
plt.show()
```

```python
sns.set(rc={'figure.figsize':(25,20)})
sns.lineplot(x="date",y="total_cases",data=india_cases)
plt.show()
```
