# **Extrapolatory Data Analysis and Feature Engineering**

#### GETTING INSIDE ABOUT THE LENGTH OF SENTENCES FOR EACH REVIEW FOR ORIGINAAL DATA

```python
final_data["Length"]= final_data['Review'].str.len()
final_data.head()
```

#### WORDS PER REVIEW

```python
final_data["Word_count"] = final_data['Review'].str.split().str.len()
final_data
```

### Pre-processing

### Conversion to string type

```python
final_data['Review'] = final_data['Review'].astype(str)
final_data
```

### Converting to lower case

```python
final_data['Review'] = final_data['Review'].apply(lambda x: " ".join(x.lower() for x in x.split()))
final_data['Review'].head()
```

### Removing special characters

```python
final_data['Review'] = final_data['Review'].str.replace('[^\w\s]', '')
final_data['Review']
```

```python
final_data['Review']  = final_data['Review'] .str.replace('[#,@,&,!,=,(,)]', '',regex=True)
final_data['Review'] 
```

### Stopwords Removal

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import stopwords
Stop_words = stopwords.words('english')
Stop_words[0:10]
```

```python
final_data["Number of Stopwords"] = final_data['Review'].apply(lambda x: [w for w in str(x).lower().split() if w in stopwords.words('english')])
final_data
```

Now we got how many stopwords are there in each review. Now we need to remove them.


```python
final_data['Review'] = final_data['Review'].apply(lambda x: " ".join(x for x in x.split() if x not in Stop_words))
final_data['Review']
```

### Spelling Correction

```python
from textblob import TextBlob, Word, Blobber
final_data["Review"].apply(lambda x: str(TextBlob(x).correct()))
final_data["Review"]
```

## Categorizing as positive and neutral

```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
corpus=[]

for i in range(0, 1002):
  review = re.sub('[^a-zA-Z]', ' ', final_data['Review'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus[0:5]
```

# **Categorizing the data in positive and Neutral**

```python
from sklearn.feature_extraction.text import CountVectorizer
model = CountVectorizer(max_features = 1420)
```

```python
X = model.fit_transform(corpus).toarray()
y = final_data.iloc[:, -1].values
```

```python
import pickle
bow_path = 'BOW_Sentiment_Analyzing.pkl'
pickle.dump(model, open(bow_path, "wb"))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 2)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
# Exporting NB Classifier to later use in prediction
import joblib
joblib.dump(classifier, 'c2_Classifier_Sentiment_Model') 
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)

accuracy_score(y_test, y_pred)
```
