```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("loan.csv")
df.head()
```

```python
# We will remove Loan_ID as it is of no use to us

df.drop("Loan_ID",axis = 1,inplace = True)
df.head()
```

```python
df.info()
```

```python
# Now we will drop all the null value columns

df_new = df.dropna()
df_new.info()
```

```python
#Now let us apply label encoding because it will coonvert all object data colum to int column ad at the same time make them convert into machine laguage which can further be used in modelling
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df_new["Gender"] = label_encoder.fit_transform(df_new["Gender"])
df_new["Married"] = label_encoder.fit_transform(df_new["Married"])
df_new["Dependents"] = label_encoder.fit_transform(df_new["Dependents"])
df_new["Education"] = label_encoder.fit_transform(df_new["Education"])
df_new["Self_Employed"] = label_encoder.fit_transform(df_new["Self_Employed"])
df_new["Property_Area"] = label_encoder.fit_transform(df_new["Property_Area"])
df_new["Loan_Status"] = label_encoder.fit_transform(df_new["Loan_Status"])
df_new.head()
```

```python
#now we will save the lable encoded as a new csv file
df_new.to_csv("F:\insideAIML\df_req.csv")
```

```python
#now we will load that updated csv file
df_req = pd.read_csv("F:\insideAIML\df_req.csv")
df_req.head()
```

```python
df_req.info()
```

```python
x = df_req.drop("Loan_Status",axis = 1)
y = df_req["Loan_Status"]
print(x.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.1, random_state = 10)
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

### Applying Logistic regression

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(solver='saga')
LR.fit(x_train,y_train) #training 

y_pred1 = LR.predict(x_test)

from sklearn.metrics import f1_score
print("Logistic Regression Accuracy is : ", f1_score(y_test, y_pred1)*100,"%")
```

### Applying SVM 

```python
from sklearn import svm
svm = svm.SVC(kernel = "sigmoid") 
svm.fit(x_train,y_train)

y_pred2 = svm.predict(x_test) 

print("SVM Accuracy is: ",(f1_score(y_pred2,y_test,average = "weighted"))*100,"%")
```

### Applying Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
LoanTree = DecisionTreeClassifier(criterion="entropy", max_depth = 5)  
LoanTree.fit(x_train,y_train)

y_pred3 = LoanTree.predict(x_test)

from sklearn.metrics import f1_score
print("Decision Trees accuracy is: ",f1_score(y_test,y_pred3)*100,"%")
```

### Applying Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000,criterion="entropy")  
rf.fit(x_train,y_train)

y_pred4 = rf.predict(x_test)

from sklearn.metrics import f1_score
print("Random Forest accuracy is: ",f1_score(y_test,y_pred4)*100,"%")
```

###### We can conclude that best accuracy is shown by  Random Forest Algorithm

```python
data = pd.read_csv("F:\insideAIML\df_fifa.csv")
data.head()
```

```python
data.drop(["Unnamed: 0"], axis = 1, inplace = True)
data.head()
```

### Apply the correlation and remove the column which are more than 60% correlated

```python
cor_matrix = data.corr().abs()
cor_matrix
```

```python
# creating upper triangular matrix and then comparing and getting more than 60% correlated column
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.60)]
print(to_drop)
```

This is the list of all columns which needs to be removed because they are more than 60% correlated.

```python
data.drop(to_drop, axis = 1,inplace = True)
data.head()
```

## Apply the variance and remove the columns whose variance is less that 20%

```python
from sklearn.feature_selection import VarianceThreshold
var_thres = VarianceThreshold(threshold=0.20)
var_thres.fit(data)
```

```python
var_thres.get_support()
```

```python
constant_columns = data.columns[var_thres.get_support() == False]
constant_columns
```

```python
data.drop(constant_columns, axis=1, inplace = True)
data.head()
```
